```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def binarize(img, Tth):
    T = np.mean(img) 
    # print(T)
    Tprev = T + 1 
    while abs(T - Tprev) >= Tth: 
        Tprev = T
        # print(img[img < T])
        G1 = img[img > T] 
        # print(G1)
        G2 = img[img <= T] 
        # print(G2)
        u1 = np.mean(G1) 
        # print(u1)
        u2 = np.mean(G2) 
        T = 0.5 * (u1 + u2) 
        # print(T)
    return (img > T).astype(np.uint8) * 255 

# Load the image
img = cv2.imread('../DATA/apple.jfif', cv2.IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


img = np.array(img)

thresholded = binarize(img, Tth=0.01)

# cv2.imwrite('thresholded.png', thresholded)
plt.imshow(thresholded)
# print(img)
```
