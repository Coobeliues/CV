```python
import cv2
import os

# Создаем папку "cuted", если ее нет
if not os.path.exists('Akhmedov Sanzhar'):
    os.makedirs('Akhmedov Sanzhar')

# Загружаем каскад Хаара для обнаружения лиц
face_cascade = cv2.CascadeClassifier('../../DATA/haarcascades/haarcascade_frontalface_default.xml')

# Перебираем каждый файл в папке "images"
for filename in os.listdir('../../фото/train/Akhmedov Sanzhar/'):
    # Загружаем изображение, которое необходимо обрезать
    img = cv2.imread(os.path.join('../../фото/train/Akhmedov Sanzhar/', filename))

    # Преобразуем изображение в оттенки серого для повышения эффективности обработки
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Находим лица на изображении
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    # Счетчик для подсчета сохраненных обрезанных лиц
    count = 0

    # Проходимся по каждому обнаруженному лицу и обрезаем его
    for (x, y, w, h) in faces:
        # Если мы уже сохранили 50 обрезанных лиц, выходим из цикла
        if count == 45:
            break

        face = img[y:y+h, x:x+w]
        # Сохраняем обрезанное изображение лица в папку "cuted" с уникальным именем файла
        cv2.imwrite(os.path.join('Akhmedov Sanzhar', 'face_{}_{}'.format(filename[:-4], count) + '.jpg'), face)

        # Увеличиваем счетчик сохраненных обрезанных лиц
        count += 1
```

```python
import cv2
import os
from PIL import Image
import numpy as np


# Загрузка каскада Хаара для обнаружения лиц
face_cascade = cv2.CascadeClassifier('../../DATA/haarcascades/haarcascade_frontalface_default.xml')

# Папка с исходными изображениями
input_folder = '../../фото/test/Zhumabekov Magjan/'

# Папка для сохранения обрезанных изображений
output_folder = 'Zhumabekov Magjan'

# Создаем выходную папку, если она не существует
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Проходимся по всем изображениям в папке
for filename in os.listdir(input_folder):
    # Загрузка изображения
    # img = cv2.imread(os.path.join(input_folder, filename))
    img = Image.open(os.path.join(input_folder, filename))
    # print(filename)
    img_array = np.asarray(img)

    # Преобразуем изображение в оттенки серого для повышения эффективности обработки
    gray = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
    # Находим лица на изображении
    faces = face_cascade.detectMultiScale(gray, 1.2, 5)
    # Обрезаем изображения и сохраняем их в новую папку
    for (x, y, w, h) in faces:
        face_img = gray[y:y+h, x:x+w]
        face_resized = cv2.resize(face_img, (150, 150))
        cv2.imwrite(os.path.join(output_folder, 'cropped_' + filename), face_resized)
```
