```python
import cv2
import numpy as np
import math 

drawing = False
mode = True
ix, iy = -1, -1
color_g = (0,255, 0)
color_r = (0,0,255)
colr = True

def draw_shape(event,x,y,flags,param):
    global ix,iy,drawing,mode, colr, color_g, color_r
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            if colr == True:
                if mode == True:
                    r = int(math.sqrt( ((ix-x)**2)+((iy-y)**2)) )
                    cv2.circle(img,(x,y),r,color_r,-1)
                else:
                    cv2.rectangle(img, (ix,iy),(x,y),color_r,-1)
            else:
                if mode == True:
                    r = int(math.sqrt( ((ix-x)**2)+((iy-y)**2)) )
                    cv2.circle(img,(x,y),r,color_g,-1)
                else:
                    cv2.rectangle(img, (ix,iy),(x,y),color_g,-1)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if colr == True:
            if mode == True:
                r = int(math.sqrt( ((ix-x)**2)+((iy-y)**2)))
                cv2.circle(img,(x,y),r,color_r,-1)
            else:
                cv2.rectangle(img, (ix,iy),(x,y),color_r,-1)
        else:
            if mode == True:
                r = int(math.sqrt( ((ix-x)**2)+((iy-y)**2)) )
                cv2.circle(img,(x,y),r,color_g,-1)
            else:
                cv2.rectangle(img, (ix,iy),(x,y),color_g,-1)
    
    
cv2.namedWindow(winname='frame')

cv2.setMouseCallback('frame', draw_shape)

###### SHOWING IMAGE WITH OPENCV #####

img = np.zeros((750,750,3), np.uint8)


        
while(1):
    cv2.imshow('frame',img)
    k = cv2.waitKey(1) & 0xFF
    if k == ord('1'):
        mode = not mode
    elif k == ord('2'):
        mode = True
    elif k == ord('3'):
        colr = not colr
    elif k == ord('4'):
        colr == True
    elif k == 27:
        break
        
cv2.destroyAllWindows()
```
