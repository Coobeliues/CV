```python
import numpy as np
import matplotlib.pyplot as plt
import cv2


lower_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_lowerbody.xml')
upper_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_upperbody.xml')
body_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_fullbody.xml')
video = cv2.VideoCapture('../DATA/2_human.mp4')


width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
writer = cv2.VideoWriter('../DATA/hum_det2wweeq.mp4', cv2.VideoWriter_fourcc(*'XVID'),25, (width, height))

if(video.isOpened()==False):
    print('Error Reading Video')
    

while True:
    ret,frame = video.read()    
    norm_video = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    bodies = body_cascade.detectMultiScale(norm_video)
    upper = upper_cascade.detectMultiScale(norm_video)
    lower = lower_cascade.detectMultiScale(norm_video)
    
    for (x,y,w,h) in bodies:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        # roi_gray = norm_video[y:y+h, x:x+w]
        # roi_color = frame[y:y+h, x:x+w]  
    for (x,y,w,h) in upper:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
    #     roi_gray = norm_video[y:y+h, x:x+w]
    #     roi_color = frame[y:y+h, x:x+w]  
    for (x,y,w,h) in lower:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
    #     roi_gray = norm_video[y:y+h, x:x+w]
    #     roi_color = frame[y:y+h, x:x+w]  
         
    if ret == True:
        
        writer.write(frame)

        cv2.imshow('video', frame)

        if cv2.waitKey(25) & 0xFF == ord("q"):
            break
    else:
        break

video.release()
writer.release()
cv2.destroyAllWindows() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

body_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_fullbody1.xml')
video = cv2.VideoCapture('../DATA/full_bodies.avi')


width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
writer = cv2.VideoWriter('../DATA/hum_det2wweeq.mp4', cv2.VideoWriter_fourcc(*'XVID'),25, (width, height))

if(video.isOpened()==False):
    print('Error Reading Video')
    

while True:
    ret,frame = video.read()    
    norm_video = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    bodies = body_cascade.detectMultiScale(norm_video)
    
    for (x,y,w,h) in bodies:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,0),2)
    
    if ret == True:
        
        writer.write(frame)

        cv2.imshow('video', frame)

        if cv2.waitKey(25) & 0xFF == ord("q"):
            break
    else:
        break

video.release()
writer.release()
cv2.destroyAllWindows() 
```
