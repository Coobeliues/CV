```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

cap = cv2.VideoCapture(0)

eye_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml')
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer = cv2.VideoWriter('../DATA/student_capture_hw3_eyes_nose.mp4', cv2.VideoWriter_fourcc(*'XVID'),25, (width, height))

# def detect_and_blur_plate(img):
#     roi = img[y:y+h,x:x+w] 
#     # final = cv2.medianBlur(roi, 7)
#     final = np.power(roi,5)
#     l = img
#     s = final
#     l[y:y+h, x:x+w] = s


# def detect_plate(img):
#     global x,w,y,h
#     detect_img = img.copy()
    
#     number_rects = haarcascade.detectMultiScale(detect_img)
    
#     for (x,y,w,h) in number_rects:
#         cv2.rectangle(detect_img, (x,y), (x+w,y+h),(255,0,0),3)
        
        
#     return detect_img


def detect_eye(img):
    
    face_img = img.copy()
    
    eyes = eye_cascade.detectMultiScale(face_img)
    
    detect_img = img.copy()
    l = face_img
    for i, number_rect in enumerate(eyes):
        x, y, w, h = number_rect
        if i <= 2:
            roi = img[y:y+h,x:x+w] 
            final = np.power(roi,5)
            s = final
            l[y:y+h, x:x+w] = s
        else:
            roi = img[y:y+h,x:x+w] 
            # final = cv2.medianBlur(roi, 7)
            final = np.power(roi,0)
            # l = img
            s = final
            l[y:y+h, x:x+w] = s
    
    # for (x,y,w,h) in eyes:
    #     cv2.rectangle(face_img, (x,y), (x+w,y+h),(255,255,255),10)
    #     roi = img[y:y+h,x:x+w] 
    #     # final = cv2.medianBlur(roi, 7)
    #     final = np.power(roi,5)
    #     l = img
    #     s = final
    #     l[y:y+h, x:x+w] = s
        
    return face_img



cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read(0)
    
    # frame = detect_face(frame)
    frame = detect_eye(frame)
    writer.write(frame)
    cv2.imshow('Face Detection', frame)
    
    c = cv2.waitKey(1)
    if c == 27:
        break
        
cap.release()
writer.release()
cv2.destroyAllWindows()
    
```
