```python
#1
import cv2


def ask_for_tracker():
    print("Welcome! What Tracker API would you like to use?")
    print("Enter 0 for BOOSTING: ")
    print("Enter 1 for MIL: ")
    print("Enter 2 for KCF: ")
    print("Enter 3 for TLD: ")
    print("Enter 4 for MEDIANFLOW: ")
    print("Enter 5 for CSRT: ")
    print("Enter 7 for MOSSE: ")
    
    choice = input("Please select your tracker: ")
    
    if choice == '0':
        tracker = cv2.TrackerBoosting_create()
    if choice == '1':
        tracker = cv2.TrackerMIL_create()
    if choice == '2':
        tracker = cv2.TrackerKCF_create()
    if choice == '3':
        tracker = cv2.TrackerTLD_create()
    if choice == '4':
        tracker = cv2.TrackerMedianFlow_create()
    if choice == '5':
        tracker = cv2.TrackerCSRT_create()
    if choice == '7':
        tracker = cv2.TrackerMOSSE_create()

    return tracker


tracker = ask_for_tracker()
tracker_name = str(tracker).split()[0][1:]

cap = cv2.VideoCapture('../DATA/full_bodies.avi')
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
writer = cv2.VideoWriter('../DATA/people_trace_hw3.mp4', cv2.VideoWriter_fourcc(*'VIDX'), 30, (width,height))

ret,frame = cap.read()

roi = cv2.selectROI(frame, False)
ret = tracker.init(frame, roi)

list_x, list_y = [], []

while True:
    ret, frame = cap.read()
        
    success, roi = tracker.update(frame)
    (x, y, w, h) = tuple(map(int, roi))
    if success:
        p1, p2 = (x, y), (x + w, y + h)
        list_x.append(x + w//2)
        list_y.append(y + h//2)
        
        cv2.rectangle(frame, p1, p2, (255,0,0), 3)

    for i in range(1,len(list_x)):
        cv2.line(frame,(list_x[i-1], list_y[i-1]), (list_x[i], list_y[i]), (0,0,255), 4)
    
    writer.write(frame)
    
    cv2.imshow(tracker_name, frame)

    if cv2.waitKey(25) & 0xFF == ord("q"):
        break

        
cap.release()
writer.release()
cv2.destroyAllWindows()
```
