```python
import os
import random
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Список поддерживаемых форматов
supported_formats = ['.jpg', '.jpeg', '.png', '.bmp']

# Создание списка папок, содержащих изображения
folders = ['фото/train/Akhmedov Sanzhar', 'фото/train/Alikhan Zhamankhan', 'фото/train/Baibatchayeva Anar', 'фото/train/Dalabay Abylay', 
           'фото/train/Daniyar Raiymbekov', 'фото/train/Duisenov Alikhan', 'фото/train/Ernar', 'фото/train/Kenzhebay Askar', 
           'фото/train/Kyran Yerkin' , 'фото/train/Maxat Ualikhan', 'фото/train/Rysmambetov Nurtugan', 'фото/train/Sagynbek Nurbolat', 
           'фото/train/Yerkin Yernar', 'фото/train/Zhetessov Nur', 'фото/train/Zhumabekov Magjan']



# Определение параметров аугментации
params = {
    'rotation_range': 45,
    'brightness_range': (0.5, 1.5),
    'shear_range': 30,
    'zoom_range': (0.5, 1.0),
    'horizontal_flip': True,
    'vertical_flip': True,
    'channel_shift_range': 20,
    'fill_mode': 'nearest',
    'cval': 0
}

# Цикл по каждой папке с изображениями
for folder in folders:
    print(f"Augmenting images in folder: {folder}")
    # Получаем список файлов в папке
    files = os.listdir(folder)
    # Цикл по каждому файлу
    for file in files:
        # Получаем путь к файлу
        file_path = os.path.join(folder, file)
        # Получаем расширение файла
        file_ext = os.path.splitext(file_path)[1].lower()
        # Проверяем, поддерживается ли формат файла
        if file_ext in supported_formats:
            # Загружаем изображение
            img = cv2.imread(file_path)
            # Преобразуем изображение в массив NumPy
            img = np.array(img)
            # Добавляем новую размерность для корректной работы генератора
            img = np.expand_dims(img, axis=0)
            # Создаем генератор для аугментации изображения
            datagen = ImageDataGenerator(**params)
            # Генерируем новые изображения
            augmented_images = datagen.flow(img, batch_size=1, save_prefix='aug', save_format='jpg')
            # Цикл по каждому сгенерированному изображению
            for i, new_image in enumerate(augmented_images):
                # Если было сгенерировано более 10 изображений, выходим из цикла
                if i > 9:
                    break
                # Создаем новое имя файла с учетом индекса сгенерированного изображения
                new_file_name = f"{os.path.splitext(file)[0]}_aug_{i}{file_ext}"
                # Создаем путь для сохранения нового изображения
                new_file_path = os.path.join(folder, new_file_name)
                # Сохраняем новое изображение
                cv2.imwrite(new_file_path, new_image[0])
```

```python
def create_dir(dir, children):
    new_dir = dir / children
    new_dir.mkdir()
    return new_dir

base_dir = create_dir(Path('.'), 'dataset')

train_dir = create_dir(base_dir, 'train')
test_dir = create_dir(base_dir, 'test')
valid_dir = create_dir(base_dir, 'validation')


###############
train_dogs_dir = create_dir(train_dir, 'dogs')
train_cats_dir = create_dir(train_dir, 'cats')

test_dogs_dir = create_dir(test_dir, 'dogs')
test_cats_dir = create_dir(test_dir, 'cats')

valid_dogs_dir = create_dir(valid_dir, 'dogs')
valid_cats_dir = create_dir(valid_dir, 'cats')


########
import shutil

def copy_files(src, dest, name, start, stop):
    fnames = [f'{name}.{i}.jpg' for i in range(start,stop)]
    for fname in fnames:
        src_file = src / fname
        dest_file = dest / fname
        shutil.copy(src_file, dest_file)
        
        
        
########
src = Path('.') / 'train' / 'train'

copy_files(src, train_dogs_dir, 'dog', 0, 1000)
copy_files(src, train_cats_dir, 'cat', 0, 1000)

copy_files(src, test_dogs_dir, 'dog', 1000, 1500)
copy_files(src, test_cats_dir, 'cat', 1000, 1500)

copy_files(src, valid_dogs_dir, 'dog', 1500, 2000)
copy_files(src, valid_cats_dir, 'cat', 1500, 2000)




#########
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu',
                  input_shape=(150,150,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(512, 'relu'),
    layers.Dense(1, 'sigmoid') # 1 нейрон – кошка или собака
])






###########################
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)


###########
def image_generator(dategen, dir):
    return dategen.flow_from_directory(
        dir,
        target_size=(150,150),
        batch_size=20,
        class_mode='binary'
    )

train_generator = image_generator(train_datagen, train_dir)
test_generator = image_generator(test_datagen, test_dir)
valid_generator = image_generator(valid_datagen, valid_dir)



##############
from datetime import datetime
from tensorflow.keras.callbacks import TensorBoard

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

log_dir = "logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

model.fit(
    train_generator,
    steps_per_epoch=100,
    epochs=30,
    validation_data=valid_generator, 
    callbacks=[tensorboard_callback])




# Преобразование данных: 
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=.2,
    height_shift_range=.2,
    shear_range=.2,
    zoom_range=.2,
    horizontal_flip=True,
)



#####
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

img_path = train_cats_dir / 'cat.1.jpg'
img = image.load_img(img_path, target_size=(150,150))
x = image.img_to_array(img)
x = x.reshape((1,) + x.shape) # изменение формы на (1,150,150,3)
i = 0
for batch in train_datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(image.array_to_img(batch[0]))
    i += 1
    if i == 2:
        break

plt.show()



###########
model = models.Sequential([
    ## CЛОИ CNN

    layers.Dropout(0.5), # Отключит нейрон с вероятностью 0.5
    layers.Flatten(),
    layers.Dense(512, 'relu'),
    layers.Dense(1, 'sigmoid')
])

train_generator = image_generator(train_datagen, train_dir)
model.fit(
    train_generator,
    steps_per_epoch=100,
    epochs=30,
    validation_data=valid_generator, 
    callbacks=[tensorboard_callback])
```
